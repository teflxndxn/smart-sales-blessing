```python
import findspark
findspark.init()

from pyspark.sql import SparkSession
from pyspark.sql import functions as F
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Start Spark Session
spark = SparkSession.builder \
    .appName("SmartSales") \
    .config("spark.jars", "/Users/stephneyaganaga/Desktop/school-assessment/smart-sales-blessing/lib/sqlite-jdbc-3.49.1.0.jar") \
    .getOrCreate()

# Load Tables
df_sales = spark.read.format("jdbc") \
    .option("url", "jdbc:sqlite:/Users/stephneyaganaga/Desktop/school-assessment/smart-sales-blessing/data/dw/smart_sales.db") \
    .option("dbtable", "sale") \
    .option("driver", "org.sqlite.JDBC") \
    .load()

df_customer = spark.read.format("jdbc") \
    .option("url", "jdbc:sqlite:/Users/stephneyaganaga/Desktop/school-assessment/smart-sales-blessing/data/dw/smart_sales.db") \
    .option("dbtable", "customer") \
    .option("driver", "org.sqlite.JDBC") \
    .load()

df_products = spark.read.format("jdbc") \
    .option("url", "jdbc:sqlite:/Users/stephneyaganaga/Desktop/school-assessment/smart-sales-blessing/data/dw/smart_sales.db") \
    .option("dbtable", "product") \
    .option("driver", "org.sqlite.JDBC") \
    .load()

# Register Views for SQL
df_sales.createOrReplaceTempView("sale")
df_products.createOrReplaceTempView("product")
df_customer.createOrReplaceTempView("customer")

# 💡 Spark: Total Sales by Category
df_sales_with_category = spark.sql("""
SELECT s.*, p.category
FROM sale s
JOIN product p ON s.product_id = p.product_id
""")

df_sales_with_time = df_sales_with_category.withColumn(
    "sale_date_parsed", F.to_date(F.col("sale_date"), "M/d/yyyy")
).withColumn(
    "year", F.year(F.col("sale_date_parsed"))
).withColumn(
    "month", F.month(F.col("sale_date_parsed"))
).withColumn(
    "quarter", F.quarter(F.col("sale_date_parsed"))
)


result = df_sales_with_time.groupBy("category", "month").sum("sale_amount")
result.show()
result_pd = result.toPandas()

# 🎨 Plot: Total Sales by Product Category
plt.figure(figsize=(10, 6))
sns.barplot(data=result_pd, x="category", y="sum(sale_amount)")
plt.xticks(rotation=45)
plt.title("Total Sales by Product Category")
plt.ylabel("Total Sales")
plt.xlabel("Product Category")
plt.tight_layout()
plt.show()

# 💡 Pandas: Sales Growth Over Time
import sqlite3

# Connect directly with SQLite using pandas
connection = sqlite3.connect("/Users/stephneyaganaga/Desktop/school-assessment/smart-sales-blessing/data/dw/smart_sales.db")
query = "SELECT sale_date, sale_amount FROM sale ORDER BY sale_date;"
data = pd.read_sql_query(query, connection)
connection.close()

# Parse date & prepare for time series plot
data['sale_date'] = pd.to_datetime(data['sale_date'])
data['month'] = data['sale_date'].dt.to_period('M')

# Monthly aggregation
monthly_sales = data.groupby('month')['sale_amount'].sum().reset_index()
monthly_sales['month'] = monthly_sales['month'].dt.to_timestamp()

plt.figure(figsize=(10, 6))
plt.plot(monthly_sales['month'], monthly_sales['sale_amount'], marker='o', color='blue')
plt.title('Sales Growth Over Time', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Total Sales', fontsize=14)
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

monthly_sales['rolling_avg'] = monthly_sales['sale_amount'].rolling(window=3).mean()

plt.figure(figsize=(10,6))
plt.plot(monthly_sales['month'], monthly_sales['sale_amount'], marker='o', label='Actual Sales', color='blue')
plt.plot(monthly_sales['month'], monthly_sales['rolling_avg'], linestyle='--', label='3-Month Rolling Avg', color='red')
plt.title('Sales Growth with 3-Month Rolling Average', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Total Sales', fontsize=14)
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# Prepare daily data
data['day'] = data['sale_date'].dt.day
data['month_str'] = data['sale_date'].dt.strftime('%Y-%m')

# Pivot table
heatmap_data = data.pivot_table(index='day', columns='month_str', values='sale_amount', aggfunc='sum')

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap="YlGnBu", linewidths=0.5)
plt.title('Sales Heatmap: Day vs Month', fontsize=16)
plt.xlabel('Month')
plt.ylabel('Day of Month')
plt.show()
```
