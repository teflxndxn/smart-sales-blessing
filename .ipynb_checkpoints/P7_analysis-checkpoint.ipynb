```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pyspark.sql import SparkSession
from pyspark.sql import functions as F

# Initialize Spark session
spark = SparkSession.builder \
    .appName("Sales Analysis") \
    .getOrCreate()

# File paths (adjust if needed)
sales_csv = "/Users/stephneyaganaga/Desktop/school-assessment/smart-sales-blessing/data/prepared/sales_data_prepared.csv"
products_csv = "/Users/stephneyaganaga/Desktop/school-assessment/smart-sales-blessing/data/prepared/products_data_prepared.csv"
customers_csv = "/Users/stephneyaganaga/Desktop/school-assessment/smart-sales-blessing/data/prepared/customers_data_prepared.csv"

# Load data
df_sales = spark.read.csv(sales_csv, header=True, inferSchema=True)
df_products = spark.read.csv(products_csv, header=True, inferSchema=True)
df_customer = spark.read.csv(customers_csv, header=True, inferSchema=True)

# Join and transform data
df_full = df_sales.join(df_products, df_sales.ProductID == df_products.ProductID, "inner") \
    .join(df_customer, df_sales.CustomerID == df_customer.CustomerID, "inner") \
    .withColumn("sale_date_parsed", F.to_date("SaleDate", "M/d/yyyy")) \
    .withColumn("month", F.month("sale_date_parsed")) \
    .withColumn("year", F.year("sale_date_parsed")) \
    .withColumn("revenue", F.col("SaleAmount") * F.col("UnitPrice")) \
    .withColumn("profitability", F.col("revenue") * F.lit(0.2))  # Example profit margin

# Aggregation
df_aggregated = df_full.groupBy("ProductName", "month", "Region").agg(
    F.sum("SaleAmount").alias("total_sales_volume"),
    F.avg("UnitPrice").alias("average_unit_price"),
    F.sum("revenue").alias("total_revenue")
)

# Visualization - Total Revenue by Product and Month
plt.figure(figsize=(12, 6))
sns.barplot(data=df_pd, x="ProductName", y="total_revenue", hue="month")
plt.xticks(rotation=45)
plt.title("Total Revenue by Product and Month")
plt.savefig("visuals/filename.png", bbox_inches='tight')
plt.show()

# Visualization - Total Sales Volume by Product and Month (Changed to Line Plot)
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_pd, x="ProductName", y="total_sales_volume", hue="month", marker="o")
plt.xticks(rotation=45)
plt.title("Total Sales Volume by Product and Month (Line Plot)")
plt.savefig("visuals/filename.png", bbox_inches='tight')
plt.show()

# Visualization - Average Unit Price by Product and Month
plt.figure(figsize=(12, 6))
sns.barplot(data=df_pd, x="ProductName", y="average_unit_price", hue="month")
plt.xticks(rotation=45)
plt.title("Average Unit Price by Product and Month")
plt.savefig("visuals/filename.png", bbox_inches='tight')
plt.show()

```
